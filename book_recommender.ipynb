```python
ratings_with_book = ratings.merge(book, on = "ISBN")
ratings_with_book.head(3)
```

```python
num_rating = ratings_with_book.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head(5)
```

```python
num_rating.rename(columns={'rating' : 'num_of_rating'},inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_book.merge(num_rating,on = 'title')
final_rating.head()
```

```python
final_rating=final_rating[final_rating['num_of_rating']>50]
final_rating.head()
```

```python
final_rating.sample(10)
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

```python
book_pivot = book_pivot.sub(book_pivot.mean(axis=1), axis=0).fillna(0)
```

```python
book_pivot
```

######### including new library scipy

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=100, random_state=42)
book_reduced = svd.fit_transform(book_sparse)
print("Before SVD:", book_sparse.shape)
print("After SVD:", book_reduced.shape)
```

```python
from sklearn.neighbors import NearestNeighbors as nn
```

```python
model = nn(metric='cosine', algorithm='brute')
```

```python
model.fit(book_reduced)
```

```python
# Transform this book’s vector using the SVD model before finding neighbors
book_vector = svd.transform(book_pivot.iloc[237,:].values.reshape(1,-1))
distance, suggestion = model.kneighbors(book_vector, n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    book_vector = svd.transform(book_pivot.iloc[book_id, :].values.reshape(1, -1))
    distance, suggestion = model.kneighbors(book_vector, n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)

```

```python
book_name= 'Hemlock Bay'

recommend_book(book_name)
```
